### üó£Ô∏èThis notebook will explain how to run open source model via ollama

##### ‚öôÔ∏è Set up of Open source model `ollama`
1. Download ollama from the website URL: https://ollama.com/
2. Install ollama like other software
3. Open PowerShell and run the following command to check ollama installation
    - `ollama`

4. Installed required model in ollama. Run the following command in PowerShell.
    - `ollama run <<model_name>>` - This will download the model if it is not present in the local system. Here is some examples
    - `ollama run phi3`
    - `ollama run gemma3:270m`

5. To start the ollama service open terminal and run `ollama serve`

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
MODEL_NAME = "llama3.2"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
message = [{"role":"user", "content":"What is the capital of USA?"}]
respone = ollama.chat.completions.create(model=MODEL_NAME, messages=message )
```
